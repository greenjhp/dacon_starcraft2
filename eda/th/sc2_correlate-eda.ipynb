```python
df = df[df['event'].isin(['Selection'])]
arr_units = df['event_contents'].map(units.findall)
del df['event_contents']
df = pd.concat([df, arr_units], axis=1)
```

```python
unit_dict = {}
species_units = {'T':{}, 'Z':{}, 'P':{}}
for idx, data in df.iterrows():
    if not data['game_id'] in unit_dict:
        unit_dict[data['game_id']] = [{}, {}]
    for unit_name, unit_id in data['event_contents']:
        if unit_name and len(unit_name) > 0: # and len(npcs.findall(unit_name)) == 0:
            # check the enemy has it
            if not unit_id in unit_dict[data['game_id']][(data['player'] + 1 ) % 2]:
                unit_dict[data['game_id']][data['player']][unit_id] = unit_name
                # save unit names per each species
                if unit_name in species_units[data['species']]:
                    species_units[data['species']][unit_name] += 1
                else:
                    species_units[data['species']][unit_name] = 1
        else:
            print(f"{idx}: [gid {data['game_id']}] Invalid item name {unit_name} ({unit_id})")
```

```python
for sp_key in ['Z', 'P', 'T']:
    print('-------------------------------', sp_key)
    print(pd.Series(species_units[sp_key]).sort_values(ascending=False)[100:])
```

```python
npc_unit_names = []
for zerg_key in species_units['Z']:
    if zerg_key in species_units['P'] and zerg_key in species_units['T']:
        npc_unit_names.append(zerg_key)
npc_unit_names
```

```python
for game_id in unit_dict:
    for npc_unit in npc_unit_names:
        for player_id in range(2):
            name_id_lookup = {unit_dict[game_id][player_id][k]:k for k in unit_dict[game_id][player_id]}
            if npc_unit in name_id_lookup:
                unit_id = name_id_lookup[npc_unit]
                del unit_dict[game_id][player_id][unit_id]
                print(f'{npc_unit}({unit_id}) deleted in game {game_id} for player {player_id}')
```
